```python
import numpy as np
acc_32 = np.load("./acc_32_642-649.npy",allow_pickle=True).item()
acc_8 = np.load("./acc_8_642-649.npy",allow_pickle=True).item()
```

```python
import matplotlib.pyplot as plt
acc_32_list = []
acc_8_list = []
for k in acc_32.keys():
    if((acc_32[k]-acc_8[k])>2):
        print( k,acc_32[k],acc_32[k]-acc_8[k])
    acc_32_list.append(acc_32[k])
    acc_8_list.append(acc_8[k])
plt.scatter(acc_32_list,acc_8_list)
plt.plot(range(82,95),range(82,95),'--',c="r")
plt.xlabel("float32 acc")
plt.ylabel("int8 acc")
plt.title("The acc comparision between float32 and int8")
plt.show()
    
```
